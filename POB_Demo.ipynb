```python
#Define Psi=((2/L)^(1/2))*sin(n*pi/L)

def psi(x,n,L): 
  return np.sqrt(2.0/L)*np.sin(float(n)*np.pi*x/L)
```

```python
h=6.62607e-34    #planck's constant in joules
me=9.1093837e-31  # mass of an electron in kg

# Defining a function to compute the energy

# Energy for (n,l,m) = (h^2/(8*mL^2)
# 1 J = 6.242*10^18 eV 
# 1 M = 1.*10^10 Angstrons
# 1J = kg*m^2*s^(-2)
#using length units is Angstroms and then converted into electron volts
# (Js)^2/(8kg*m^2)=J
# (h**2 / (me*8))* (1e10)**2 *6.242e+18  is the prefactor 
# to convert Ang^2 to m^2 and Joules to eV

def En(n,L,me): return (h**2 / (me*8))* (1e10)**2 *6.242e+18*((float(n)/L)**2)

# Here the users inputs the value of L
L = float(input("Enter the value of L (in Angstroms) = "))
nmax = int(input("Enter the maximum value of n you want to plot = "))

for n in range(1,nmax+1):
  Energy=En(n,L,me)
  print(n,Energy,"eV")
```
