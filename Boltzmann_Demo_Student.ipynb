```python
import matplotlib as mpl # matplotlib library for plotting and visualization
import matplotlib.pylab as plt # matplotlib library for plotting and visualization
import numpy as np #numpy library for numerical manipulation, especially suited for data arrays
from numpy import exp
import warnings
warnings.filterwarnings('ignore')
```

```python
#In Kcal/mol
#kT=RT/N_a
#R=k*N_a (Kcal/(mol*K)
kb=0.001985875

def populations_T(states,deltaE,degenercy,T):
  States=np.linspace(0,states-1,states)
  E_levels=deltaE*States
  boltz_weights=exp(-E_levels/(kb*T))
  q=np.sum(boltz_weights)
  population=boltz_weights/q
  return(States,population)

def populations_T_degen(states,deltaE,degen,T):
  States=np.linspace(0,states-1,states)
  E_levels=deltaE*States
  boltz_weights=exp(-E_levels/(kb*T))*degen
  q=np.sum(boltz_weights)
  population=boltz_weights/q
  return(States,population)
```

We are going to examine some preoperties of the Boltzmann Distribution.  We are going to explore how temperature, energy level spacing, and degenercy affect the populations of molecular energy levels.

To review, the ratio of the populations between two molecular energy levels is governed by the **Boltzmann Population Ratio**. where $N_i$
is the population of state i, $N_j$ is the popuulation of state j, ${\epsilon}_i$ is the energy of state i and ${\epsilon}_j$ is the energy of state j. 


$\LARGE \frac{N_i}{N_j} =\frac{e^{-\beta\epsilon_i}}{e^{-\beta\epsilon_j}}=e^{-\beta(\epsilon_j-\epsilon_i)}$ 

$\LARGE \beta=\frac{1}{kT}$

First we will consider two energy levels, can you find an energy level spacing (in Kcal/mol) where the upper energy is population to ~10% of the lower energy level?  ~50% of the lower energy level?

```python
#In Kcal/mol
#R=k*N_a (kcal/(mol*K)
kb=0.001985875

n = int(input("Enter number of energy levels = "))
dE = float(input("Enter the Difference between energy levels (in kcal/mol) = "))
T = float(input("Temperature (K) = "))

plt.clf()
s,p=populations_T(n,dE,0,T)
plt.bar(s,p)
plt.xlabel("State",size=20)
plt.ylabel("Population",size=20)
plt.xticks(s,size=20)
plt.yticks(size=20)
plt.title(r'$\Delta$E=%s kcal/mol, T=%s K'%(dE,T),size=20)

plt.show()
```

Now we can generalize to systems with multiple energy levels.  Here the populations of each state are determined by the **Boltzmann Distribution** and we introduce the **partition function** *q*

$\LARGE \frac{N_i}{N} =\frac{e^{-\beta\epsilon_i}}{q} $


$\LARGE q= \sum_{i}^ne^{-\beta\epsilon_i} $

where q is the sum of the boltzmann weights over $n$ states

No we consider the effect of the degenercy of each energy level.  

If two states $i$ and $j$ have degenercies $g_i$ and $g_j$ their populations are given by:

$\LARGE \frac{N_i}{N_j} =\frac{g_i}{g_j}\frac{e^{-\beta\epsilon_i}}{e^{-\beta\epsilon_j}}=\frac{g_i}{g_j}e^{-\beta(\epsilon_j-\epsilon_i)}$ 

and the boltzmann distribution becomes:

$\LARGE \frac{N_i}{N} =\frac{g_ie^{-\beta\epsilon_i}}{q} $


$\LARGE q= \sum_{i}^ng_ie^{-\beta\epsilon_i} $
